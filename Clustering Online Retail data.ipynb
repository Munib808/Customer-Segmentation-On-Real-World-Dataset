# *We will have separate analysis of outliers we don't directly exclude them as they show the customers that do shopping frequently and spend more than others this means that outliers indicate important customers so we separate the outliers and ooutlier-free data so that we can analyze them separately*

```python
M_Q1 = aggregated_df["MonetaryValue"].quantile(0.25)
M_Q3 = aggregated_df["MonetaryValue"].quantile(0.75)
M_IQR = M_Q3 - M_Q1

monetary_outlier_df = aggregated_df[(aggregated_df["MonetaryValue"] > (M_Q3 + 1.5 * M_IQR)) | (aggregated_df["MonetaryValue"] < (M_Q1 - 1.5 * M_IQR))].copy()
monetary_outlier_df.describe()
```

```python
F_Q1 = aggregated_df["Frequency"].quantile(0.25)
F_Q3 = aggregated_df["Frequency"].quantile(0.75)
F_IQR = F_Q3 - F_Q1

frequency_outlier_df = aggregated_df[(aggregated_df["Frequency"] > (F_Q3 + 1.5 * F_IQR)) | (aggregated_df["Frequency"] < (F_Q1 - 1.5 * F_IQR))].copy()
frequency_outlier_df.describe()
```

```python
non_outlier_df = aggregated_df[(~aggregated_df.index.isin(monetary_outlier_df.index)) & (~aggregated_df.index.isin(frequency_outlier_df.index))]

non_outlier_df.describe()
```

# **K-Means Clustering**

```python
k_values = range(2, 12)

inertia_values = []
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42,max_iter = 1000)
    
    y_kmeans = kmeans.fit_predict(scaled_data_df)    
    silhouette_scores.append(silhouette_score(scaled_data_df, y_kmeans))
    inertia_values.append(kmeans.inertia_)


plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.plot(k_values, inertia_values, marker='o')
plt.title('Elbow Method: Inertia vs. k')
plt.xlabel('Number of Clusters (k)')
plt.xticks(range(2, 13))
plt.ylabel('Inertia')
plt.grid(True)


plt.subplot(1, 3, 2)
plt.plot(k_values, silhouette_scores, marker='o')
plt.title('Silhouette Score vs. k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 13))
plt.grid(True)

plt.tight_layout()
plt.show()
```

# **The elbow is around k = 4 or 5**

```python
full_clustering_df = pd.concat([non_outlier_df,outlier_cluster_df])
full_clustering_df
```

```python
cluster_labels = {
    0: "RETAIN",
    1: "RE-ENGAGE",
    2: "NURTURE",
    3: "REWARD",
    -1: "PAMPER",
    -2: "UPSELL",
    -3: "DELIGHT"
}
full_clustering_df["Cluster_labels"] = full_clustering_df["Cluster"].map(cluster_labels)
full_clustering_df
```

```python
cluster_counts = full_clustering_df['Cluster_labels'].value_counts()
feature_means = full_clustering_df.groupby('Cluster_labels')[['Recency', 'Frequency', 'MonetaryValue']].mean()

fig, ax1 = plt.subplots(figsize=(12, 8))

sns.barplot(x=cluster_counts.index, y=cluster_counts.values, ax=ax1, palette='viridis', hue=cluster_counts.index)
ax1.set_ylabel('Number of Customers', color='b')
ax1.set_title('Cluster Distribution with Average Feature Values')

ax2 = ax1.twinx()

sns.lineplot(data=feature_means, ax=ax2, palette='Set2', marker='o')
ax2.set_ylabel('Average Value', color='g')

plt.show()
```
